Importando pandas e biblotecas

```python
import pandas as pd
df = pd.read_csv('../data/asteroids_data.csv', sep=',')
pd.set_option('display.max_columns', None)
df.head()
```

**Normalizando colunas e Tratando Nulls**

```python
#Convertendo colunas strings em datetime
colunas = ['orbit_determination_date', 'first_observation_date', 'last_observation_date']
df[colunas] = df[colunas].apply(pd.to_datetime)
```

```python
#Dropar coluna sentry_data
df = df.drop('sentry_data', axis=1)

#Preenchendo nulls da coluna short_name
def fill_short(df):
    for idx, row in df.iterrows():
        if row['short_name'] == '' or pd.isna(row['short_name']):
            parts = row['name'].split(' ')
            if len(parts) >= 3:
                fill = parts[2][:-1]
                df.at[idx, 'short_name'] = fill
    return df

df = fill_short(df)

#Criando diâmetro médio
df['diameter_medium'] = (df['diameter_max_m'] + df['diameter_min_m']) / 2

df.to_csv('../data/clean_asteroids_data.csv', index=False, sep=',')
df
```

**Criando tabela de Asteroides Perigosos**

```python
#Filtrando pelo potencial de perigo e depois eliminando essa coluna
df_perigo = df[df['potentially_hazardous'] == True]
df_perigo = df_perigo.drop('potentially_hazardous', axis=1)

#Filtrando pelos maiores diâmetros e exportando
df_perigo = df_perigo.sort_values(['diameter_medium', 'min_orbit_intersection'], ascending=False).reset_index(drop=True)
df_perigo.to_csv('../data/dangest_asteroids.csv', index=False, sep=',')
display(df_perigo)
```

**Fazendo gráfico dos 10 asteroides mais perigosos**

```python
import plotly.express as px
top10 = df_perigo.iloc[0:10]
fig = px.bar(top10, x='diameter_medium', y='short_name', orientation='h', title='10 Asteroides Mais Perigosos', color='diameter_medium', labels={'short_name': 'Nome', 'diameter_medium': 'Diâmetro Médio'}, height=800)
fig.show()
fig.write_image('../graphic/10_Asteroids_Dangerous.png', height=800, width=1920)
```

Quantidade de observações ao longo do tempo

```python
#Nova coluna com apenas o ano da primeira observação
df['observation_year'] = df['first_observation_date'].dt.year
#Cotando as observações por ano
df_observation = df.groupby('observation_year')['id'].count().reset_index()
#Renomeando coluna
df_observation = df_observation.rename(columns={'id': 'amount'})
df_observation.to_csv('../data/observation_year.csv')
df_observation
```

```python
fig_observation = px.line(df_observation, x='observation_year', y='amount', title='Observações De Asteroides Por Ano', labels={'observation_year': 'Ano', 'amount': 'Quantidade'})
fig_observation.show()
fig_observation.write_image('../graphic/Observation_Asteroids_perYear.jpg', height=800, width=1920)
```

**Média de diâmetro e magnitude por potencial de perigo**

```python
df_asteroids_year = df.groupby('potentially_hazardous').agg(diameter_medium=('diameter_max_m', 'mean'), magnitude_medium=('magnitude', 'mean')).reset_index()
df_asteroids_year[['diameter_medium', 'magnitude_medium']] = df_asteroids_year[['diameter_medium', 'magnitude_medium']].round(2)
df_asteroids_year
```
